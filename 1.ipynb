## Exercise 1 : College Dataset - Basic Data Processing


This exercise relates to the `College` data set, which can be found in the fle `College.csv` on the book website. It contains a number of variables for 777 diferent universities and colleges in the US. The variables are


- `Private` : Public/private indicator
- `Apps` : Number of applications received
- `Accept` : Number of applicants accepted
- `Enroll` : Number of new students enrolled
- `Top10perc` : New students from top 10 % of high school class
- `Top25perc` : New students from top 25 % of high school class
- `F.Undergrad` : Number of full-time undergraduates
- `P.Undergrad` : Number of part-time undergraduates
- `Outstate` : Out-of-state tuition
- `Room.Board` : Room and board costs
- `Books` : Estimated book costs
- `Personal` : Estimated personal spending
- `PhD` : Percent of faculty with Ph.D.s
- `Terminal` : Percent of faculty with terminal degree
- `S.F.Ratio` : Student/faculty ratio
- `perc.alumni` : Percent of alumni who donate
- `Expend` : Instructional expenditure per student
- `Grad.Rate` : Graduation rate

_d) Use the `pd.plotting.scatter_matrix()` function to produce a scatterplot matrix of the first columns `[Top10perc, Apps, Enroll]`. Recall that you can reference a list C of columns of a data frame `A` using `A[C].`_


```python
college[["Top10perc", "Apps", "Enroll"]]
```

```python
# Create the scatter matrix with customization
axes = pd.plotting.scatter_matrix(
    college[["Top10perc", "Apps", "Enroll"]],
    figsize=(10, 10),
    color="tomato",
    diagonal="kde",
    marker="o",
    alpha=0.7,
    grid=True,
)

# Rotate labels and adjust font size
for ax in axes.flatten():
    ax.yaxis.label.set_rotation(90)
    ax.tick_params(axis="both", labelsize=10)

plt.suptitle(
    "Scatter Matrix: Top 10%, Applications, and Enrollment", fontsize=16, weight="bold"
)
plt.tight_layout()
plt.subplots_adjust(top=0.93)
plt.show()
```

_e) Use the boxplot() method of college to produce side-by-side boxplots of Outstate versus Private._


```python
# Create the boxplot
ax = college.boxplot(
    column="Outstate", by="Private", grid=False, color="tomato", figsize=(8, 6)
)

# Setting the title and labels
plt.title("Out-of-State Tuition by College Type", fontsize=16, weight="bold")
plt.suptitle("")
plt.xlabel("Private College?", fontsize=12)
plt.ylabel("Out-of-State Tuition ($)", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.tight_layout()
plt.show()
```

_f) Create a new qualitative variable, called Elite, by binning the Top10perc variable into two groups based on whether or not the
proportion of students coming from the top 10% of their high school classes exceeds 50%._


```python
college["Elite"] = pd.cut(college["Top10perc"], [0, 50, 100], labels=["No", "Yes"])
```

```python
college["Elite"].value_counts()
```

```python
# Create the boxplot
ax = college.boxplot(
    column="Outstate", by="Elite", grid=False, color="blue", figsize=(8, 6)
)

# Add plot aesthetics
plt.title("Out-of-State Tuition by Elite College Status", fontsize=16, weight="bold")
plt.suptitle("")  # Remove pandas default suptitle
plt.xlabel("Elite College?", fontsize=12)
plt.ylabel("Out-of-State Tuition ($)", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.tight_layout()
plt.show()
```

_g) Use the plot.hist() method of college to produce some histograms with difering numbers of bins for a few of the quantitative variables. The command plt.subplots(2, 2) may be useful: it will divide the plot window into four regions so that four plots can be made simultaneously. By changing the arguments you can divide the screen up in other combinations._


```python
fig, ax = plt.subplots(2, 2)
fig.set_size_inches(12, 10)

# Global title
fig.suptitle("Distribution of College Features", fontsize=18, weight="bold")

# Top10perc
ax[0, 0].hist(college["Top10perc"], bins=5, color="tomato", edgecolor="black")
ax[0, 0].set_title("Top 10% of Students", fontsize=14)
ax[0, 0].set_xlabel("Top 10 Percentage", fontsize=12)
ax[0, 0].set_ylabel("Count of Colleges", fontsize=12)
ax[0, 0].grid(axis="y", linestyle="--", alpha=0.5)

# Top25perc
ax[0, 1].hist(college["Top25perc"], bins=5, color="lime", edgecolor="black")
ax[0, 1].set_title("Top 25% of Students", fontsize=14)
ax[0, 1].set_xlabel("Top 25 Percentage", fontsize=12)
ax[0, 1].set_ylabel("Count of Colleges", fontsize=12)
ax[0, 1].grid(axis="y", linestyle="--", alpha=0.5)

# Books
ax[1, 0].hist(college["Books"], bins=5, color="orange", edgecolor="black")
ax[1, 0].set_title("Book Costs", fontsize=14)
ax[1, 0].set_xlabel("Books ($)", fontsize=12)
ax[1, 0].set_ylabel("Count of Colleges", fontsize=12)
ax[1, 0].grid(axis="y", linestyle="--", alpha=0.5)

# PhD
ax[1, 1].hist(college["PhD"], bins=5, color="skyblue", edgecolor="black")
ax[1, 1].set_title("Faculty with PhDs", fontsize=14)
ax[1, 1].set_xlabel("PhD Percentage", fontsize=12)
ax[1, 1].set_ylabel("Count of Colleges", fontsize=12)
ax[1, 1].grid(axis="y", linestyle="--", alpha=0.5)

# Layout adjustments
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Reserve space for suptitle
plt.show()
```

_h) Continue exploring the data, and provide a brief summary of what you discover._


```python
# Correlation matrix heatmap

plt.figure(figsize=(12, 10))
sns.heatmap(college.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Numeric College Features")
plt.show()
```

```python
# Scatter plot: Top 10% vs Graduation Rate
plt.figure(figsize=(6, 5))
plt.scatter(
    college["Top10perc"], college["Grad.Rate"], color="mediumseagreen", alpha=0.6
)
plt.xlabel("Top 10% of Incoming Class")
plt.ylabel("Graduation Rate")
plt.title("Top 10% vs Graduation Rate")
plt.grid(True)
plt.tight_layout()
plt.show()

# Scatter plot: Top 10% vs Applications
plt.figure(figsize=(6, 5))
plt.scatter(college["Top10perc"], college["Apps"], color="coral", alpha=0.6)
plt.xlabel("Top 10% of Incoming Class")
plt.ylabel("Number of Applications")
plt.title("Top 10% vs Applications Received")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Summary :

- **Top 10% vs Graduation Rate:** A positive correlation is observed — colleges with more students from the top 10% of their high school class tend to have higher graduation rates.

- **Top 10% vs Applications:** Colleges attracting more top-performing students also receive a higher number of applications, suggesting higher popularity or prestige.


```python
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Outstate Tuition
college.boxplot(column="Outstate", by="Private", ax=ax[0], color="skyblue")
ax[0].set_title("Out-of-State Tuition")
ax[0].set_xlabel("Private College")
ax[0].set_ylabel("Tuition ($)")
ax[0].grid(axis="y", linestyle="--", alpha=0.5)

# Enrollment
college.boxplot(column="Enroll", by="Private", ax=ax[1], color="lime")
ax[1].set_title("Student Enrollment")
ax[1].set_xlabel("Private College")
ax[1].set_ylabel("Enrolled Students")
ax[1].grid(axis="y", linestyle="--", alpha=0.5)

# Graduation Rate
college.boxplot(column="Grad.Rate", by="Private", ax=ax[2], color="plum")
ax[2].set_title("Graduation Rate")
ax[2].set_xlabel("Private College")
ax[2].set_ylabel("Graduation Rate (%)")
ax[2].grid(axis="y", linestyle="--", alpha=0.5)

fig.suptitle("Private vs Public Colleges", fontsize=16, weight="bold")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

### Summary :

- **Tuition:** Private colleges charge significantly higher out-of-state tuition compared to public institutions.

- **Enrollment:** Public colleges tend to enroll more students, indicating larger institutions.

- **Graduation Rate:** Graduation rates are slightly higher on average in private colleges, potentially due to smaller class sizes and greater student support.


```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# PhD vs Grad.Rate
ax[0].scatter(college["PhD"], college["Grad.Rate"], color="dodgerblue", alpha=0.6)
ax[0].set_title("Faculty with PhDs vs Graduation Rate")
ax[0].set_xlabel("PhD Percentage")
ax[0].set_ylabel("Graduation Rate")
ax[0].grid(True)

# Terminal vs Grad.Rate
ax[1].scatter(college["Terminal"], college["Grad.Rate"], color="darkorange", alpha=0.6)
ax[1].set_title("Faculty with Terminal Degrees vs Graduation Rate")
ax[1].set_xlabel("Terminal Degree Percentage")
ax[1].set_ylabel("Graduation Rate")
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

### Summary :

- **PhD and Terminal Degree Percentages:** Colleges with a higher percentage of faculty holding PhDs or terminal degrees generally have better graduation rates. This suggests that faculty qualifications may play a role in academic outcomes.


```python
# Create new columns
college["AcceptanceRate"] = college["Accept"] / college["Apps"]
college["YieldRate"] = college["Enroll"] / college["Accept"]

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Acceptance Rate vs Graduation Rate
ax[0].scatter(college["AcceptanceRate"], college["Grad.Rate"], alpha=0.6, color="teal")
ax[0].set_title("Acceptance Rate vs Graduation Rate", fontsize=14)
ax[0].set_xlabel("Acceptance Rate", fontsize=12)
ax[0].set_ylabel("Graduation Rate", fontsize=12)
ax[0].grid(True)

# Yield Rate vs Graduation Rate
ax[1].scatter(college["YieldRate"], college["Grad.Rate"], alpha=0.6, color="purple")
ax[1].set_title("Yield Rate vs Graduation Rate", fontsize=14)
ax[1].set_xlabel("Yield Rate", fontsize=12)
ax[1].set_ylabel("Graduation Rate", fontsize=12)
ax[1].grid(True)

plt.tight_layout()
plt.show()

# Boxplots: Private vs Public - Acceptance & Yield
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
college.boxplot(column="AcceptanceRate", by="Private", ax=ax[0], color="skyblue")
ax[0].set_title("Acceptance Rate by College Type")
ax[0].set_xlabel("Private College?")
ax[0].set_ylabel("Acceptance Rate")
ax[0].grid(axis="y", linestyle="--", alpha=0.5)

college.boxplot(column="YieldRate", by="Private", ax=ax[1], color="lime")
ax[1].set_title("Yield Rate by College Type")
ax[1].set_xlabel("Private College?")
ax[1].set_ylabel("Yield Rate")
ax[1].grid(axis="y", linestyle="--", alpha=0.5)

plt.suptitle("Admission Selectivity & Enrollment Yield", fontsize=16, weight="bold")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Compute total estimated cost (tuition + room + books + personal)
college["TotalCost"] = (
    college["Outstate"] + college["Room.Board"] + college["Books"] + college["Personal"]
)

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Tuition vs Instructional Expenditure
ax[0].scatter(college["Outstate"], college["Expend"], alpha=0.6, color="coral")
ax[0].set_title("Out-of-State Tuition vs Instructional Spending", fontsize=14)
ax[0].set_xlabel("Outstate Tuition ($)", fontsize=12)
ax[0].set_ylabel("Expenditure per Student ($)", fontsize=12)
ax[0].grid(True)

# Total Cost vs Graduation Rate
ax[1].scatter(college["TotalCost"], college["Grad.Rate"], alpha=0.6, color="olive")
ax[1].set_title("Total Cost vs Graduation Rate", fontsize=14)
ax[1].set_xlabel("Total Estimated Cost ($)", fontsize=12)
ax[1].set_ylabel("Graduation Rate (%)", fontsize=12)
ax[1].grid(True)

plt.tight_layout()
plt.show()
```
